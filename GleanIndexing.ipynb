### Create Content

```python
import os
import random
import string
from docx import Document

# Define functions to generate random text and create Word documents from a basic set of words.
def generate_random_text(length):
    words = [
        "the", "fast", "mountain biker", "rides", "jumps", "over", "logs", "tires",
        "and", "changes", "leads", "from", "the", "wins", "in", "the", "suspension",
        "with", "great", "speed", "and", "endurance", "while", "the", "sun", "course"
    ]
    punctuation = ['.', '!', ',', ';', ':']
    text = []
    current_length = 0

    while current_length < length:
        word = random.choice(words)
        text.append(word)
        current_length += len(word) + 1  # +1 for the space

        # Add punctuation occasionally
        if current_length < length and random.random() < 0.1:
            text.append(random.choice(punctuation))
            current_length += 1

    # Join the words into a single string
    return ' '.join(text)[:length]

def create_word_document(filepath, text):
    doc = Document()
    doc.add_paragraph(text)
    doc.save(filepath)

# Store the generated documents in this directory. 
directory = "/Users/barkz/Desktop/Glean/docs"

# Check if the directory exists, if not create it.
if not os.path.exists(directory):
    os.makedirs(directory)

# Create 50 random Word documents with 500 characters each.
for i in range(1, 51):
    random_text = generate_random_text(500)
    filename = f"random_document_{i}.docx"
    filepath = os.path.join(directory, filename)
    create_word_document(filepath, random_text)
    print(f"Created {filepath}")

print("All documents created successfully in the directory: " + directory)
```

### Read Content and Create Document Dictionary

There are 3 object_types created in the Interview Datasource (interviewds). 
* 50 documents, "STORM MTB Riders TX", assigned as "EngineeringDoc". Random content word generation.
* 50 documents, "STORM MTB Riders SATX", assigned "Assignment". Random content word generation.
* 50 documents, "Mountain Bike Story", assigned "Document". Basic dictionary based content generation.

```python
import os
from docx import Document

# Specify the directory containing the documents.
directory = "/Users/barkz/Desktop/Glean/docs"

# Create an empty list to store the document data as defined in the Glean API.
documents_data = []

# Iterate over the files in the directory.
for i, filename in enumerate(os.listdir(directory), start=1):
    filepath = os.path.join(directory, filename)
    
    # Read file.
    if filename.endswith(".docx"):
        doc = Document(filepath)
        text_content = "\n".join([para.text for para in doc.paragraphs])
        mime_type = "application/vnd.openxmlformats-officedocument.wordprocessingml.document"
    else:
        continue  # Skip non-docx files
    
    # Create document data dictionary.
    document_data = {
        "datasource": "interviewds",
        "object_type": "document", 
        "title": f"Mounatin Bike Story {i}",
        "id": f"mtb-story-{i}",
        "view_url": f"https://app.glean.com/",
        "body": {
            "mime_type": mime_type,
            "text_content": text_content
        },
        "permissions": {
            "allow_anonymous_access": True
        },
        "collection": "TX Storm Riders Collection"
    }
    
    # Add the document data to the list
    documents_data.append(document_data)
```

### Index Documents

```python
from glean_indexing_api_client.api import documents_api
from glean_indexing_api_client.model.index_document_request import IndexDocumentRequest
from glean_indexing_api_client.model.document_definition import DocumentDefinition
from glean_indexing_api_client.model.content_definition import ContentDefinition
from glean_indexing_api_client.model.user_reference_definition import (
    UserReferenceDefinition,
)
from glean_indexing_api_client.model.document_permissions_definition import (
    DocumentPermissionsDefinition,
)

#  List to store IndexDocumentRequest instances.
requests = []

# Iterate over the document data and create IndexDocumentRequest instances.
for doc_data in documents_data:
    request = IndexDocumentRequest(
        document=DocumentDefinition(
            datasource=doc_data["datasource"],
            object_type=doc_data["object_type"],
            title=doc_data["title"],
            id=doc_data["id"],
            view_url=doc_data["view_url"], # <-- Used app.glean.com as a placeholder.
            body=ContentDefinition(
                mime_type=doc_data["body"]["mime_type"],
                text_content=doc_data["body"]["text_content"]
            ),
            permissions=DocumentPermissionsDefinition(
                allow_anonymous_access=doc_data["permissions"]["allow_anonymous_access"]
            ),
        )
    )
    requests.append(request)

# Create an instance of Glean DocumentsApi.
documents_api = documents_api.DocumentsApi(api_client)

# Iterate over the IndexDocumentRequest instances and call the indexdocument_post method.
for index, request in enumerate(requests, start=1):
    document_name = request.document.title
    try:
        documents_api.indexdocument_post(request)
        print(f"Successfully indexed {document_name}.")
    except indexing_api.ApiException as e:
        print(f"Exception when calling DocumentsApi->indexdocument_post for Document {index}: {e}\n")
```
